# Practice Code Samples for Digital Image Processing

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

## Gray-Level Transforms

```python
img = cv2.imread("../lena.png")
img_g = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
print(img.shape, img_g.shape)
```

```python
plt.axis("off")
plt.imshow(img[...,::-1])
```

```python
plt.axis("off")
plt.imshow(img_g, cmap="gray")
```

### Log Transform

```python
img_log = np.array(np.log(img_g), dtype=np.uint8)

f, axarr = plt.subplots(1,2)
axarr[0].axis("off")
axarr[0].imshow(img_g, cmap="gray")
axarr[1].axis("off")
axarr[1].imshow(img_log, cmap="gray")
```

## Power Transform

```python
gamma = 1.5
img_pow = np.power(img_g, gamma)

f, axarr = plt.subplots(1,2)
axarr[0].axis("off")
axarr[0].imshow(img_g, cmap="gray")
axarr[1].axis("off")
axarr[1].imshow(img_pow, cmap="gray")
```

```python
gamma = 0.2
img_pow = np.power(img_g, gamma)

f, axarr = plt.subplots(1,2)
axarr[0].axis("off")
axarr[0].imshow(img_g, cmap="gray")
axarr[1].axis("off")
axarr[1].imshow(img_pow, cmap="gray")
```

## Image Enhacement

```python
print("Original")
plt.hist(img_g.ravel(),256,[0,256])
plt.show()
```

## Contrast Stretching

```python
img_stretched = (img_g - img_g.min())*255.0/(img_g.max() - img_g.min())

f, axarr = plt.subplots(1,2)
axarr[0].axis("off")
axarr[0].imshow(img_g, cmap="gray")
axarr[1].axis("off")
axarr[1].imshow(img_stretched, cmap="gray")
```

```python
plt.hist(img_stretched.ravel(),256,[0,256])
plt.show()
```

### Histogram Equaliztion

```python
img_equalized = cv2.equalizeHist(img_g)

f, axarr = plt.subplots(1,2)
axarr[0].axis("off")
axarr[0].imshow(img_g, cmap="gray")
axarr[1].axis("off")
axarr[1].imshow(img_equalized, cmap="gray")
```

```python
plt.hist(img_equalized.ravel(),256,[0,256])
plt.show()
```

```python
kernel = np.array([[-1, -1, -1],
                   [-1, 8, -1],
                   [-1, -1, -1]])
img_mask = cv2.filter2D(img_g, -1, kernel)
img_super_sharp = cv2.addWeighted(img_g,0.95,img_mask,0.05,0)

f, axarr = plt.subplots(1,3)
axarr[0].axis("off")
axarr[0].imshow(img_g, cmap="gray")
axarr[1].axis("off")
axarr[1].imshow(img_mask, cmap="gray")
axarr[2].axis("off")
axarr[2].imshow(img_super_sharp, cmap="gray")
```
