```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_lena = cv2.imread('lena.png')
img_dollar = cv2.imread('dollar.tif')
```

## Task 01

```python
def negate(img, L):
    return L - 1 - img
```

## Testing

### RGB

```python
plt.imshow(negate(img_lena, 256)[:,:,::-1])
plt.show()
```

```python
plt.imshow(negate(img_dollar, 256)[:,:,::-1])
plt.show()
```

### Grayscale

```python
img_lena_gray = cv2.cvtColor(img_lena, cv2.COLOR_BGR2GRAY)
plt.imshow(negate(img_lena_gray, 256), cmap='gray')
plt.show()
```

```python
img_dollar_gray = cv2.cvtColor(img_dollar, cv2.COLOR_BGR2GRAY)
plt.imshow(negate(img_dollar_gray, 256), cmap='gray')
plt.show()
```

### Binary

```python
img_lena_bi = img_lena_gray.copy()
img_lena_bi[ img_lena_bi <= 127 ] = 0
img_lena_bi[ img_lena_bi > 127 ] = 255

plt.imshow(negate(img_lena_bi, 256), cmap='gray')
plt.show()
```

```python
img_dollar_bi = img_dollar_gray.copy()
img_dollar_bi[ img_dollar_bi <= 127 ] = 0
img_dollar_bi[ img_dollar_bi > 127 ] = 255

plt.imshow(negate(img_dollar_bi, 256), cmap='gray')
plt.show()
```

## Task 02

```python
def gradient(img):
    
    img[img <= 127] = 0
    img[img > 127] = 1
    
    height, width = img.shape
    out = np.zeros((height, width), dtype=np.uint8)
    
    for i in range(height):
        for j in range(width):
            if j+1 < width:
                out[i,j] = abs(img[i,j+1] - img[i,j])
    return out
```

## Testing

```python
plt.imshow(gradient(img_lena_gray.copy()), cmap='gray')
plt.show()
```

```python
plt.imshow(gradient(img_dollar_gray.copy()), cmap='gray')
plt.show()
```

## Task 03

```python
def plane_n(img, plane):
    img[ (img & 2**(plane-1)) == 0 ] = 0
    img[ (img & 2**(plane-1)) != 0 ] = 1
    
    return img
```

```python
for i in range(1,9):
    print('Plane', i)
    plt.axis('off')
    plt.imshow(plane_n(img_dollar_gray.copy(), i),cmap='binary')
    plt.show()
```

```python
for i in range(1,9):
    print('Plane', i)
    plt.axis('off')
    plt.imshow(plane_n(img_lena_gray.copy(), i),cmap='binary')
    plt.show()
```
